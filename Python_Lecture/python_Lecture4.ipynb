```python
class Car :
  def __init__(self, make, model, color, price):
    self.make = make
    self.model = model
    self.color = color
    self.price = price

  def setMake(self, make): # 설정자 메소드
    self.make = make

  def getMake(self, make): # 접근자 메소드
    return self.make

  #차량에 대한 정보를 문자열로 요약하여서 반환한다.
  def getDesc(self):
    return "차량=("+str(self.make)+","+\
      str(self.model)+","+\
      str(self.color)+","+\
      str(self.price)+")"
```

```python
# Car 클래스를 상속하는 ElectricCar 클래스
class ElectricCar(Car):
  def __init__(self, make, model, color, price, batterySize):
    super().__init__(make, model, color, price)
    self.batterySize=batterySize
  
  def setBatterySize(self, batterySize): # 설정자 메소드
    self.batterySize = batterySize

  def getBatterySize(self, batterySize): # 접근자 메소드
    return self.batterySize

myCar = ElectricCar("Tisla", "ModelS", "white", 10000, 0)
myCar.setMake("Tesla")
myCar.setBatterySize(60)
print(myCar.getDesc())

```

```python
class Person: # 부모 클래스1
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def show(self):
    print(self.name, self.age)

class Student:  # 부모 클래스2
  def __init__(self,id):
    self.id = id
  
  def getId(self):
    return self.id

class CollegeStudent(Person,Student): # 부모 1,2를 상속하는 자녀클래스
  def __init__(self,name,age,id):
    Person.__init__(self,name,age)
    Student.__init__(self,id)

obj = CollegeStudent("Kim",22,"100036")
obj.show()
print(obj.getId())
```

```python
# 메소드 오버라이딩
# 부모클래스의 메소드를 다시 정의가 가능

# *파이썬에서는 "오버로딩"은 지원하지 않는다!
```

```python
# 예제 직원과 매니저

class Employee: # 부모 클래스
  def __init__(self, name, salary):
    self.name=name
    self.salary=salary

  def getSalary(self): # 급여 획득자
    return self.salary

  def __str__(self): # print 시 실행
    return f"이름: {self.name}, 월급: {self.getSalary()}"

class Manager(Employee): # 자녀 클래스
  def __init__(self,name,salary,bonus): # 자녀의 생성자
    super().__init__(name,salary) # 부모의 생성자
    self.bonus = bonus

  def getSalary(self):
    return super().getSalary() + self.bonus

kim = Manager("김철수", 2000000, 1000000) 
print(kim) # 부모 클래스의 __str__() 실행.
# self.name 과 self.getSalary가 실행
# self.getSalary는 자녀의 getSalary
```

```python
try:
  z = 1/0
except ZeroDivisionError: # 예외처리
  print("0으로 나눌 수 없습니다.")
```

```python
try:
  z = 1/0
except ZeroDivisionError as e: # 예외를 e로 지정
  print(e) # 예외를 출력
```

```python
try :
  print("예외 시작")
  raise Exception("일부로 발생") # 예외를 일부로 발생 시킨다.
  print("예외 끝") # 실행되지 않음: 예외가 발생하며 except로 이동하기 때문에
except Exception as e:
  print(e)
```
