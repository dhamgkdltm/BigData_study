```python
import pandas as pd

f = pd.read_csv("/content/wine.csv")

print(f.iloc[0,:])
```

```python
f.corr()
```

```python
X = f[['volatile acidity','residual sugar','chlorides','total sulfur dioxide','density','quality','class']] # f.corr()으로 alcohol과 상관계수가 +-0.25 이상인 컬럼들만 확인
y = f['alcohol']


X_train = X.iloc[:4000,:] # X_train은 즉 피쳐 데이터이고, 전체 데이터에서 모델 훈련에만 쓸 4000개의 데이터와 컬럼을 모두 가져온다.
X_valid = X.iloc[4000:,:] # X_valid는 즉 피쳐 데이터이고, 전체 데이터에서 모델 validation에만 쓸 4000개 이외의 데이터와 컬럼을 모두 가져온다.
y_train = y.iloc[:4000] # y_train는 훈련에 쓸 정답값 데이터, 훈련할 데이터 X_train과 매칭하여 슬라이싱. 컬럼은 1개뿐이므로 따로 인덱싱하진 않는다.
y_valid = y.iloc[4000:] # validation에 쓸 정답값 데이터, X_valid와 매칭하여 가져온다.
```

```python
# 데이터를 제대로 슬라이싱했는지 확인
print(len(X_train))
print(len(X_valid))
print(len(y_train))
print(len(y_valid))
```

```python
from sklearn.preprocessing import PolynomialFeatures # 폴리노미얼 피쳐를 만들기 위해 sklearn의 PolynomialFeatures를 임포트

poly_features = PolynomialFeatures(degree=2, include_bias=False) # 데이터(피쳐)를 n의 제곱까지 늘릴 준비.(degree=2 이고, 데이터의 컬럼이 A,B라면 A^2,B^2,AB,A,B 이렇게 된다.)

poly_features.fit(X_train) # X_train이 가지고 있는 컬럼이 몇개인지 파악. feature를 몇개까지 늘려야하는지 분석(fit이라는 단어에 속지 말 것.)

poly_X = poly_features.transform(X_train) # X_train의 컬럼은 원래 7개 였지만, PolynomialFeatures 때문에 35개로 늘어난다. 이것을 poly_X로 정의
```

```python
from sklearn.linear_model import SGDRegressor # sklearn에서 지원해주는 SGDRegressor를 가져온다.

# SGDRegressor 객체를 생성할 때 훈련 옵션을 지정. "경사하강법"을 50번 반복(max_iter), loss를 어디까지 줄일 것인지(tol=1e-3)
# 이외 옵션은 구글에서 검색, 사람들이 주로 사용하는 셋팅값으 와서 사용.
sgd_reg = SGDRegressor(max_iter=50, tol=1e-3, penalty=None, eta0=0.1, random_state=42)
sgd_reg.fit(poly_X, y_train) # poly_features를 적용한 35개의 컬럼의 poly_X와 그에 대응하는 y_train을 모델에게 주고, SGDRegressor의 w를 학습시킨다.
```

```python
import numpy as np
X_new_poly = poly_features.transform(X_valid) # PolynomialFeatures는 이미 X_train의 컬럼이 몇개이고, 몇개까지 늘려야하는지 분석했다.
# 굳이 X_valid로 다시 분석 시킬 필요가 없다.(X_valid도 X_train과 피쳐 개수가 같기 때문)
# 그래서 그냥 위에서 PolynomialFeatures 객체를 poly_features로 만든것을 그대로 가져와서 fit 시키지 않고 transform만 x_valid에 적용하면
# X_train과 똑같이 35개의 컬럼으로 바뀐다. 그것을 X_new_poly로 정의 
```

```python
# MSE를 가지고 예측값과 실제값의 차이를 측정할 것이다.
from sklearn.metrics import mean_squared_error

y_predict = sgd_reg.predict(X_new_poly) # 훈련이 완료된 sgd_reg에 (valid_set이면서 Polynomialfeatures를 적용시킨 35개 컬럼의) X_new_poly를 넣고 prediction 값을 뽑아낸다.

print(mean_squared_error(y_predict,y_valid))
```
